```python
#Data cleaning : removing the unnecessary columns and finding the missing values
df.drop(['img_link','product_link'],axis =1,inplace = True)
df.isnull().sum()
```

```python
#finding the index location of missing values
df.loc[df.rating_count.isnull()]
```

```python
#dropping the missing values
df.drop(index =[282,324],inplace =True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
#converting the string datatype into desired datatype and data standardization

df['category'] =  df.category.apply(lambda x :x.split('|')[0])
df['discounted_price'] = df.discounted_price.apply(lambda x: x.replace('‚Çπ','').replace(',','')).astype(float)
df['actual_price'] = df.actual_price.apply(lambda x :x.replace('‚Çπ','').replace(',','')).astype(float)
df['discount_percentage'] = df.discount_percentage.apply(lambda x :x.replace('%','')).astype(float)
df['rating_count'] = df.rating_count.apply(lambda x:x.replace(',',''))
df['rating_count'] = pd.to_numeric(df['rating_count'])
```

```python
#the product most purchased will have high rating_count.sorting the rating_count column
df.sort_values(by ='rating_count',ascending =False, inplace =True)
```

```python
#dropping the duplicate values based on product_id, as the product_id need to be unique.
#A product_id may have many user_id
df.drop_duplicates(subset ='product_id',keep ='first',inplace= True,ignore_index = True)
```

```python
# In order to perform analysis, relevant data set is important
dfnew = df[['product_id', 'product_name', 'category', 'discounted_price',
       'actual_price', 'discount_percentage', 'rating', 'rating_count',
       #'about_product',
             'user_id', 'user_name','review_id']] #'review_title','review_content', 'img_link', 'product_link'],]]
```

```python
#Univariate analysis using count and box plots for numerical attribute
sns.boxplot(dfnew['discount_percentage']);
```

insight:for all the products, min discount_percentage is 30 and max is 62.

```python
sns.countplot(data=dfnew, x='rating')
plt.tight_layout()
```

insight:maximum users gave rating in the range 4.44,4.34 and 4.23 are the maximum rating given by maximun users

```python
#multivariate analysis 

corrmatrix = dfnew.corr()
sns.heatmap(corrmatrix,square =True)
```

```python
#pairplot
sns.set()
sns.pairplot(dfnew,height=2.0)
plt.show()
```

insight:rating_count is high for products having more discounts.

```python
plt.figure(figsize=(20,5))
sns.histplot(data=dfnew, x='category', stat='percent')
```

Insight:Electronics,Home&kitchen and Computer&Accesories are the three top categories which have  high market percentage. 

```python
#exploring more to three categories
 
sns.catplot(data=dfnew, x= 'category', y='rating_count', kind="strip",jitter=0.2,height =5,aspect =4.0)
```

Insight:This strip plot clearly shows that for the categorical data(nominal),there is a high demand for the products under the category Electronics,Home&kitchen and Computer&Accesories
